**Day 1: Python Basics - Syntax, Variables**

```python
'''
This script tests various arithmetic operations on different data types.
It uses the eval function to dynamically evaluate expressions and handle exceptions that may arise from invalid operations.
'''

data_types = [4, 3.14, 'hi']
operators = ['*', '/', '+', '-', '*', '%', '**', '//', '==','+=', '-=']

def arithmetic_tests(a : int, b : float, c : str):
    for d_type in data_types:
        print(f"\n*=*=*=* Testing with {repr(d_type)} *=*=*=*=*")
        for i in operators:
            print(f"\n=== Operator {i} ===\n")
            
            try: 
                result = eval(f"{repr(d_type)} {i} {a}")
                print(f"{d_type} {i} {a} = {result}\n")
            except Exception as e:
                print(f"Int Test Exception: {e}")

            try: 
                result = eval(f"{repr(d_type)} {i} {b}")
                print(f"{d_type} {i} {b} = {result}\n")
            except Exception as e:
                print(f"Float Test Exception: {e}")

            try:
                result = eval(f"{repr(d_type)} {i} {c}")
                print(f"{d_type} {i} {c} = {result}\n")
            except Exception as e:
                print(f"String Test Exception: {e}")

arithmetic_tests(2, 3.14, 'hi')
```

**Day 2: Python Data Types**

```python
def discount_calculator(age, is_student, total_purchase_amount, is_tuesday):
    """
    Calculate the discount based on age, student status, and purchase amount.  
    Additional discount if it's Tuesday.
    """
    discount_eligible = (age >= 65 or is_student) and total_purchase_amount > 50
    final_price = total_purchase_amount

    if is_tuesday and discount_eligible:
        final_price *= 0.95

    return discount_eligible, final_price

# Test case for discount_calculator function
print(discount_calculator(70, False, 60, True)) 
```

**Day 3: Control Structure - Loops**

```python
import random   
'''
The code below is a simple math quiz game that generates random arithmetic problems for the user to solve.
The user can input their answer and the program will check if it's correct.
The game continues until the user types 'exit'. 
'''

print(" ********** Welcome to the math quiz game! ********** \n Type 'exit' to quit.\n")

while input != 'exit':
    num1 = random.randint(1, 10)
    num2 = random.randint(1, 10)
    operator = random.choice(['+', '-', '*', '/'])
    correct_answer = eval(f"{num1} {operator} {num2}")
    print(f"What is {num1} {operator} {num2}?") 
    user_answer = input("Your answer: ")
    if user_answer == 'exit':
        break
    try:    
        user_answer = float(user_answer)
        if user_answer == correct_answer:
            print("Correct!")
        else:
            print(f"Wrong! The correct answer is {correct_answer}.")
    except ValueError:
        print("Invalid input. Please enter a number or 'exit' to quit.")    

```

**Day 4: Control Structure - Conditional Statements**

```python
'''Write a program that interprets a list of daily step counts. For each day’s step count:

If the steps are less than 5,000, categorize as “Sedentary”.

If the steps are between 5,000 and 7,499 (inclusive), categorize as “Lightly Active”.

If the steps are between 7,500 and 9,999 (inclusive), categorize as “Moderately Active”.

If the steps are 10,000 or more, categorize as “Very Active”.

Print the day number and the corresponding activity category.'''

daily_steps = [3000, 6000, 8000, 12000, 4500, 10000, 7000]
activity_categories = []

for day, steps in enumerate(daily_steps, start=1):
    if steps < 5000:
        category = "Sedentary"
    elif 5000 <= steps <= 7499:
        category = "Lightly Active"
    elif 7500 <= steps <= 9999:
        category = "Moderately Active"
    else:
        category = "Very Active"
    
    activity_categories.append((day, category)) 

for day, category in activity_categories:
    print(f"Day {day}: {category}")

```
