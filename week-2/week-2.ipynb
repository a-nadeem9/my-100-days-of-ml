**Day 6: Linear Algebra - Vector Operations in Python**

```python
import numpy as np

x = np.array([1,3,5,7])

print("Vector v: ", x)
```

```python
x1 = np.array([3,4])
y1 = np.array([1,2])

print("Vector x1:", x1)
print("Vector y1:", y1)
```

```python
# addition
sum = x1 + y1
print("Sum: ", sum)

x_add = np.add(x1, y1)
print("Sum using np.add: ", x_add)

# subtraction
diff = x1 - y1
print("Difference: ", diff)
```

```python
# scalar multiplication
scaled = x1 * 2
print("Scaled: ", scaled)
```

```python
# magnitude and direction

magnitude = np.linalg.norm(x1)
print("Magnitude: ", magnitude)

direction = np.arctan2(x1[1], x1[0])
print("Direction: ", direction)
```

```python
# dot product
dot_product = np.dot(x1, y1)
print("Dot product: ", dot_product)
```

```python
# cross product

print(x1, y1)
x1_3d = np.append(x1, 0)
y1_3d = np.append(y1, 0)

print(x1_3d, y1_3d)

cross_product = np.cross(x1_3d, y1_3d)
print("Cross product: ", cross_product)
```

```python
# visualizing additon and subtraction
import matplotlib.pyplot as plt

# Plotting vectors x1 and y1
plt.quiver(0, 0, x1[0], x1[1], angles='xy', scale_units='xy', scale=1, color='r', label='x1')
plt.quiver(0, 0, y1[0], y1[1], angles='xy', scale_units='xy', scale=1, color='b', label='y1')

plt.quiver(0, 0, x_add[0], x_add[1], angles='xy', scale_units='xy', scale=1, color='g', label='Addition ( x1 + y1)')
plt.quiver(0, 0, diff[0], diff[1], angles='xy', scale_units='xy', scale=1, color='y', label='Subtraction ( x1 - y1)')

plt.xlim(0, 10)
plt.ylim(0, 10)
plt.grid()
plt.title('Add/Subtract Visualization')

# Show legend and plot
plt.legend()
plt.show()
```

```python
# vector norm
norm = np.linalg.norm(x1)
print("Norm: ", norm)
```

**Activity: Vector Field Visualization**

```python
import matplotlib.pyplot as plt
```

```python
# create a grid

x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)

X, Y = np.meshgrid(x, y)
```

```python
# define the vector field
U = Y
V = -X
```

```python
# plot the vector field
plt.quiver(X, Y, U, V, color='r')

plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.title("Circular Vector Field")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
plt.show()
```

**Day 7: Linear Algebra - Matrices and Matrix Operations**

```python
import numpy as np

mat_X = np.array([[5, 3, 1], [9, 6, 3], [13, 12, 11]])
mat_Y = np.array([[4, 7, 8], [22, 45, 76], [32, 24, 54]])

mat_XY = mat_X@mat_Y

mat_XY_T = mat_XY.T
mat_XY_T2 = np.transpose(mat_XY)

mat_XY_inv = np.linalg.inv(mat_XY_T)

print(mat_XY, '\n')
print(mat_XY_T, '\n')
print(mat_XY_T2, '\n')
print(mat_XY_inv)
```

```python
import sympy as sp
from sympy.abc import x
import numpy as np
import matplotlib.pyplot as plt
```

```python
function = x**2 
```

```python
derivative = sp.diff(function, x)
print(f"The derivative of f(x) = x^2 with respect to x is: {derivative}")
```

```python
value_at_x = 2
rate_of_change = derivative.subs(x, value_at_x)
print(f"The rate of change of f(x) = x^2 at x = {value_at_x} is: {rate_of_change}")
```

```python
# Define the function and derivative as lambdas
func_lambda = sp.lambdify(x, function, 'numpy')
deriv_lambda = sp.lambdify(x, derivative, 'numpy')

# Generate x values
x_vals = np.linspace(-5, 5, 100)
y_vals = func_lambda(x_vals)
tangent_line = deriv_lambda(value_at_x) * (x_vals - value_at_x) + func_lambda(value_at_x)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='f(x) = x^2')
plt.plot(x_vals, tangent_line, label='Tangent at x = 2', linestyle='dashed')
plt.scatter(value_at_x, func_lambda(value_at_x), color='red') # point of tangency
plt.title('Function and Tangent Line')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show() 
```

**Day 9: Calculus - Integrals, Fundamental Theorems, and Applications in Machine Learning**

```python
import sympy as sp
from sympy.abc import x 
```

```python
function = x**2
```

```python
indefinite_integral = sp.integrate(function, x)

print(f"The indefinite integral of f(x) = x^2 is: {indefinite_integral}")
```

```python
# Calculate the definite integral of `f(x) = x^2` from `1` to `3`:
a, b = 1, 3
definite_integral = sp.integrate(function, (x, a, b))

print(f"The definite integral of f(x) = x^2 from {a} to {b} is: {definite_integral}")
```

```python
# Graph the function and the area under the curve between `x = 1` and `x = 3`:
import numpy as np
import matplotlib.pyplot as plt

# Define the function and integral as lambda functions
func_lambda = sp.lambdify(x, function, 'numpy')
integral_lambda = sp.lambdify(x, indefinite_integral, 'numpy')

# Generate x values
x_vals = np.linspace(0, 4, 100)
y_vals = func_lambda(x_vals)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='f(x) = x^2', color='blue')
plt.fill_between(x_vals, y_vals, where=[(x_val >= a and x_val <= b) for x_val in x_vals], color='gray', alpha=0.5)
plt.title('Function and Area Under Curve')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
