**Day 6: Linear Algebra - Vector Operations in Python**

```python
import numpy as np

x = np.array([1,3,5,7])

print("Vector v: ", x)
```

```python
x1 = np.array([3,4])
y1 = np.array([1,2])

print("Vector x1:", x1)
print("Vector y1:", y1)
```

```python
# addition
sum = x1 + y1
print("Sum: ", sum)

x_add = np.add(x1, y1)
print("Sum using np.add: ", x_add)

# subtraction
diff = x1 - y1
print("Difference: ", diff)
```

```python
# scalar multiplication
scaled = x1 * 2
print("Scaled: ", scaled)
```

```python
# magnitude and direction

magnitude = np.linalg.norm(x1)
print("Magnitude: ", magnitude)

direction = np.arctan2(x1[1], x1[0])
print("Direction: ", direction)
```

```python
# dot product
dot_product = np.dot(x1, y1)
print("Dot product: ", dot_product)
```

```python
# cross product

print(x1, y1)
x1_3d = np.append(x1, 0)
y1_3d = np.append(y1, 0)

print(x1_3d, y1_3d)

cross_product = np.cross(x1_3d, y1_3d)
print("Cross product: ", cross_product)
```

```python
# visualizing additon and subtraction
import matplotlib.pyplot as plt

# Plotting vectors x1 and y1
plt.quiver(0, 0, x1[0], x1[1], angles='xy', scale_units='xy', scale=1, color='r', label='x1')
plt.quiver(0, 0, y1[0], y1[1], angles='xy', scale_units='xy', scale=1, color='b', label='y1')

plt.quiver(0, 0, x_add[0], x_add[1], angles='xy', scale_units='xy', scale=1, color='g', label='Addition ( x1 + y1)')
plt.quiver(0, 0, diff[0], diff[1], angles='xy', scale_units='xy', scale=1, color='y', label='Subtraction ( x1 - y1)')

plt.xlim(0, 10)
plt.ylim(0, 10)
plt.grid()
plt.title('Add/Subtract Visualization')

# Show legend and plot
plt.legend()
plt.show()
```

```python
# vector norm
norm = np.linalg.norm(x1)
print("Norm: ", norm)
```

**Activity: Vector Field Visualization**

```python
import matplotlib.pyplot as plt
```

```python
# create a grid

x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)

X, Y = np.meshgrid(x, y)
```

```python
# define the vector field
U = Y
V = -X
```

```python
# plot the vector field
plt.quiver(X, Y, U, V, color='r')

plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.title("Circular Vector Field")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
plt.show()
```

**Day 7: Linear Algebra - Matrices and Matrix Operations**

```python
import numpy as np

mat_X = np.array([[5, 3, 1], [9, 6, 3], [13, 12, 11]])
mat_Y = np.array([[4, 7, 8], [22, 45, 76], [32, 24, 54]])

mat_XY = mat_X@mat_Y

mat_XY_T = mat_XY.T
mat_XY_T2 = np.transpose(mat_XY)

mat_XY_inv = np.linalg.inv(mat_XY_T)

print(mat_XY, '\n')
print(mat_XY_T, '\n')
print(mat_XY_T2, '\n')
print(mat_XY_inv)
```
