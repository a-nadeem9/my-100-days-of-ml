**Day 1: Python Basics - Syntax, Variables**

```python
data_types = [4, 3.14, 'hi']
operators = ['*', '/', '+', '-', '*', '%', '**', '//', '==','+=', '-=']

def arithmetic_tests(a : int, b : float, c : str):
    for d_type in data_types:
        print(f"\n*=*=*=* Testing with {repr(d_type)} *=*=*=*=*")
        for i in operators:
            print(f"\n=== Operator {i} ===\n")
            
            try: 
                result = eval(f"{repr(d_type)} {i} {a}")
                print(f"{d_type} {i} {a} = {result}\n")
            except Exception as e:
                print(f"Int Test Exception: {e}")

            try: 
                result = eval(f"{repr(d_type)} {i} {b}")
                print(f"{d_type} {i} {b} = {result}\n")
            except Exception as e:
                print(f"Float Test Exception: {e}")

            try:
                result = eval(f"{repr(d_type)} {i} {c}")
                print(f"{d_type} {i} {c} = {result}\n")
            except Exception as e:
                print(f"String Test Exception: {e}")

arithmetic_tests(2, 3.14, 'hi')
```
